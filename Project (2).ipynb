## Mozart Markov Chain
Creating the transition matrix

```python
import pandas as pd
import numpy as np
import json
```

```python
with open('content/mozart-lenozze-di-figaro-overture.json', 'r') as f:
  data = json.loads(f.read())


print(data)
```

```python
df_nested_list = pd.json_normalize(data)
display(df_nested_list['tracks'])
```

```python
noteData = json.loads(pd.Series(df_nested_list['tracks']).to_json())
```

```python
noteData = json.loads(pd.Series(noteData['0']).to_json())
```

```python
for i in noteData['1']['notes']:
   display(i['name'])
```

```python
notelist =[]
for i in noteData['1']['notes']:
   notelist.append(i['name'])
```

```python
#print(notelist)
df = pd.DataFrame(notelist)

# create a new column with data shifted one space
df['shift'] = df[0].shift(-1)

# add a count column (for group by function)
df['count'] = 1

# groupby and then unstack, fill the zeros
trans_mat = df.groupby([0, 'shift']).count().unstack().fillna(0)

# normalise by occurences and save values to get transition matrix
trans_mat = trans_mat.div(trans_mat.sum(axis=1), axis=0).values
names = ['A#4', 'A#5', 'A4', 'A5', 'B3', 'B4', 'B5', 'C#4', 'C#5', 'C#6', 'C5', 'C6', 'D#5',
 'D4', 'D5', 'D6', 'E4', 'E5', 'E6', 'F#4', 'F#5', 'F#6', 'F4', 'F5', 'G#4', 'G#5',
 'G4', 'G5', 'G6']
print(trans_mat)
Mat = pd.DataFrame(trans_mat, index=names, columns=names)
Mat.to_csv('transitionMatrix.csv', index=True, header=True, sep = ',')
```
